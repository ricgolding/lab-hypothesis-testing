**Challenge 1**

In this challenge, we will be working with pokemon data. The data can be found here:

- https://raw.githubusercontent.com/data-bootcamp-v4/data/main/pokemon.csv

```python
#libraries
import pandas as pd
import scipy.stats as st
import numpy as np

```

```python
df = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/pokemon.csv")
df
```

- We posit that Pokemons of type Dragon have, on average, more HP stats than Grass. Choose the propper test and, with 5% significance, comment your findings.

```python
#code here

df_dragon_hp = df[df['Type 1'] == 'Dragon']['HP']
df_grass_hp = df[df['Type 1'] == 'Grass']['HP']

#H0: mu df_dragon_hp <= df_grass_hp
#H1: mu df_dragon_hp > df_grass_hp

alpha = 0.05

stat, p_value = st.ttest_ind(df_dragon_hp,df_grass_hp,equal_var=False,alternative='greater')

if p_value > alpha:
    print("We are not able to reject the null hypothesis")

else:
    print(f"We reject the null hypothesis because alpha {alpha} is > than p_value {p_value}")
```

- We posit that Legendary Pokemons have different stats (HP, Attack, Defense, Sp.Atk, Sp.Def, Speed) when comparing with Non-Legendary. Choose the propper test and, with 5% significance, comment your findings.


```python
#code here

#H0: mu df_legendary = df_non_legendary
#H1: mu df_legendary != df_non_legendary

stat_hp, p_hp = st.f_oneway(df[df['Legendary'] == True]['HP'], df[df['Legendary'] == False]['HP'])
stat_attack, p_attack = st.f_oneway(df[df['Legendary'] == True]['Attack'], df[df['Legendary'] == False]['Attack'])
stat_defense, p_defense = st.f_oneway(df[df['Legendary'] == True]['Defense'], df[df['Legendary'] == False]['Defense'])
stat_sptatk, p_spatk = st.f_oneway(df[df['Legendary'] == True]['Sp. Atk'], df[df['Legendary'] == False]['Sp. Atk'])
stat_spdef, p_spdef = st.f_oneway(df[df['Legendary'] == True]['Sp. Def'], df[df['Legendary'] == False]['Sp. Def'])
stat_speed, p_speed = st.f_oneway(df[df['Legendary'] == True]['Speed'], df[df['Legendary'] == False]['Speed'])

p_value_dic = {'P_Value_HP':p_hp,
               'P_Value_Attack':p_attack,
               'P_Value_Defense':p_defense,
               'P_Value_Sp.Atk':p_spatk,
               'P_Value_Sp.Def':p_spdef,
               'P_Value_Speed':p_speed
              }

for key, p_value in p_value_dic.items():
    if p_value > alpha:
        print("We are not able to reject the null hypothesis")
    
    else:
        print(f"We reject the null hypothesis for {key} because alpha {alpha} is > than p_value {p_value}")        
```

**Challenge 2**

In this challenge, we will be working with california-housing data. The data can be found here:
- https://raw.githubusercontent.com/data-bootcamp-v4/data/main/california_housing.csv

```python
df = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/california_housing.csv")
df.head()
```

**We posit that houses close to either a school or a hospital are more expensive.**

- School coordinates (-118, 34)
- Hospital coordinates (-122, 37)

We consider a house (neighborhood) to be close to a school or hospital if the distance is lower than 0.50.

Hint:
- Write a function to calculate euclidean distance from each house (neighborhood) to the school and to the hospital.
- Divide your dataset into houses close and far from either a hospital or school.
- Choose the propper test and, with 5% significance, comment your findings.
 

```python
#H0: mu house_price_close <= mu house_price_far
#H1: mu house_price_close > mu house_price_far


school_long, school_lat = -118,34
hosp_long, hosp_lat = -122,37

def calc_euclidian(x,y):
    
    school_dist = np.sqrt((x - school_long) ** 2 + (y - school_lat) ** 2)
    hosp_dist = np.sqrt((x - hosp_long) ** 2 + (y - hosp_lat) ** 2)
    return school_dist < 0.50 or hosp_dist < 0.50

df['close_to_school_or_hospital'] = df.apply(lambda x: calc_euclidian(x['longitude'], x['latitude']),axis=1)


school_hospital_nearby = df[df['close_to_school_or_hospital'] == True]['median_house_value']
school_hospital_not_nearby = df[df['close_to_school_or_hospital'] == False]['median_house_value']

stat, p_value = st.ttest_ind(school_hospital_nearby,school_hospital_not_nearby,alternative='greater')

if p_value > alpha:
    print("We are not able to reject the null hypothesis")

else:
    print(f"We reject the null hypothesis because alpha {alpha} is > than p_value {p_value}")
```
